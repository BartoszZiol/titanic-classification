# Titanic - Exploratory Data Analysis (EDA)
## Cel analizy
W tej analizie eksploracyjnej przeanalizujemy dane pasażerów Titanica,
aby zrozumieć ich strukturę, znaleźć wzorce oraz przygotować je do dalszego przetwarzania.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# Wczytanie danych
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# Podstawowe informacje
print('Informacje o zbiorze treningowym:')
print(train_df.info())
print('\nInformacje o zbiorze testowym:')
print(test_df.info())
```

## Podstawowe statystyki opisowe
Poniżej przedstawiamy statystyki opisowe dla zbiorów danych.

```python
print(train_df.describe())
print(test_df.describe())
```

## Wizualizacja brakujących wartości
Analiza brakujących wartości pomoże nam określić, które cechy wymagają uzupełnienia lub usunięcia.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(train_df.isnull(), cbar=False, cmap='viridis')
plt.title('Brakujące wartości w train.csv')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(test_df.isnull(), cbar=False, cmap='viridis')
plt.title('Brakujące wartości w test.csv')
plt.show()
```

## Rozkład zmiennej docelowej
Sprawdzenie, ile osób przeżyło, a ile nie.

```python
sns.countplot(x='Survived', data=train_df)
plt.title('Liczba osób, które przeżyły / nie przeżyły')
plt.show()
```

## Rozkład wieku pasażerów
Zobaczmy, jak wygląda podział wieku pasażerów na Titanicu.

```python
sns.histplot(train_df['Age'].dropna(), kde=True, bins=30)
plt.title('Rozkład wieku pasażerów')
plt.show()
```

## Rozkład według płci
Czy płeć miała wpływ na przeżycie?

```python
sns.countplot(x='Sex', hue='Survived', data=train_df)
plt.title('Przeżycie według płci')
plt.show()
```

## Rozkład według klasy pasażera
Czy klasa podróży miała wpływ na przeżycie?

```python
sns.countplot(x='Pclass', hue='Survived', data=train_df)
plt.title('Przeżycie według klasy pasażera')
plt.show()
```

## Wnioski
- Widoczna jest duża liczba brakujących wartości w kolumnie 'Cabin'.
- Wiek pasażerów ma duży rozrzut, ale większość znajduje się w przedziale 20-40 lat.
- Kobiety miały większą szansę przeżycia.
- Pasażerowie z wyższych klas mieli większą szansę przeżycia.
